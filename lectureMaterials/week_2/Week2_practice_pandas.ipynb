### 조건이 여러개인 경우

```python
kim = data["담당"] == "김지수"
weight = data["몸무게"] > 75.5

## 두개의 조건을 합칠 때 and를 쓰면 python and와 pandas and가 충돌이 일어나서 에러.
## -> and(&)
data[kim & weight]
```

```python
## or (|)
data[kim | weight]
```

```python
# bad practice
badP = data[data['담당'] == '김지수']['몸무게']
badP
```

```python
%timeit badP
```

```python
# good practice
# data.loc[row, column]
goodP = data.loc[data['담당'] == '김지수', '몸무게']
goodP
```

```python
%timeit goodP
```

```python
data["지점"] = data["지점"].replace("강남구", "강남")
data
```

```python
data['몸무게'] = data['몸무게'].replace(3, 0)
```

```python
data
```

```python
data['몸무게'] = data['몸무게'].fillna(75.8)
## 빈값을 채우기 - fillna()
```

```python
data
```

```python
data['측정일'] = pd.to_datetime(data['측정일'])
# pd.to_datetime 을 안할 경우, 날짜를 입력했더라도 문자열로 인식하기 때문에 에러 발생

# date, year, month, day, week, weekday
data['측정일'].dt.weekday
```

# 컬럼 추가 & 수정하기

```python
# 컬럼 생성시 모두 동일한 값을 넣어줄 때
data['지역'] = '서울'
data
```

```python
data['결제'] = pd.Series(True, index=list(range(data.shape[0]+1)), dtype='Bool')
data
```

```python
data["과체중"] = data["몸무게"] > 75.7
data
```

```python
data.loc[data['몸무게'] > 75.7, "상태"] = "과체중"
data.loc[data['과체중'] != True, "상태"] = '정상'
data
```
