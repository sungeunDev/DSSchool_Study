```python
import pandas as pd
import seaborn as sns
%matplotlib inline

train = pd.read_csv('~/Git/DSSchool_Study/lectureMaterials/sample_data_titanic/train.csv')
test = pd.read_csv('~/Git/DSSchool_Study/lectureMaterials/sample_data_titanic/test.csv')
sampleSubmission = pd.read_csv('~/Git/DSSchool_Study/lectureMaterials/sample_data_titanic/gender_submission.csv')
```

# Explore
### - countplot

```python
sns.countplot(data=train, x="Embarked", hue="Survived")
```

```python
sns.countplot(data=train, x="Pclass", hue="Survived")
```

### - barplot

```python
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

- 검은색 선은 '신뢰구간'. 데이터의 편차가 어느정도인지를 보여준다.
- 선이 길수록 편차가 큰 것. (아웃라이어를 파악할 수 있음)
- 1등급은 가격차이가 많이 남. 3등급은 가격 차이가 적음.

```python
sns.countplot(data=train, x="Fare")
```

- 구간이 많이 나뉘는 column은 countplot으로 그리기에는 부적합

### - pointplot

```python
sns.pointplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

- pointplot은 기간별로 볼 때, 시계열 데이터를 시각화할 때 주로 사용

### - lmplot

```python
sns.lmplot(data=train, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

- 회기분석, regression analysis

```python
low_fare = train[train.Fare < 200]
sns.lmplot(data=low_fare, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

- Anscombe's Quartet (시각화의 중요성)
  - 평균, 분산, 회귀선 등은 같을지라도 그래프의 분포는 다를 수 있다.
  - 회귀선이 중요하게 다뤄지지 않는 이유.
  - cf. The Datasaurus Dozen (Revolutions)

### - distplot

```python
sns.distplot(low_fare["Fare"])
```

- 정규분포(Gaussian distribution)
